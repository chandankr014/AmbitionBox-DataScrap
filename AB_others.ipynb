```python
import requests
from bs4 import BeautifulSoup
```

```python
# Define headers to mimic a browser request
headers ={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
N = 500 # no.of pages
```

```python
industry = []
employees = []
ownership = []
Longevity = []
locations = []
headquater = []
no_branches = []
```

```python
rows = []

for k in range(1, N):
    url = f'https://www.ambitionbox.com/list-of-companies?page={k}'
    response = requests.get(url, headers=headers)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    div_element = soup.find_all('div', {'class':'companyCardWrapper__interLinkingWrapper'})
    for divs in div_element:
        s = divs.get_text(strip=True)
        s2 = s.split('\n')
        # print(s2)
        r = s2[0].split(' | ')
        # print(r)
        rows.append(r)

        # scrap-2
        r = k-1
        info_parts = [part.strip() for part in rows[r]]
        # print(info_parts)
        
        for part in rows[r]:
            if "Employees" in part:
                employee_count = part
                employees.append(employee_count)
                
            elif "Public" in part or "Forbes" in part or "Conglomerate" in part:
                owner = part
                ownership.append(owner)
                
            elif "years old" in part:
                age = part
                Longevity.append(age)
    
            elif "more" in part:
                loc = part
                locations.append(loc)
                
            else:
                indust = part
                industry.append(indust)
                
            # print(part)
        
        try:
            hq, ot = locations[r].split('+')
            headquater.append(hq)
            no_branches.append(ot)
        except:
            hq = None 
            ot = None
        


    

    # break
```
