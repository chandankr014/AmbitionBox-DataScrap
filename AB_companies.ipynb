```python
import requests
from bs4 import BeautifulSoup
```

```python
# Define headers to mimic a browser request
headers ={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
N = 1000
```

```python
company__ = []
```

```python
for k in range(1, N):
    url = f'https://www.ambitionbox.com/list-of-companies?page={k}'
    response = requests.get(url, headers=headers)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all <h2> elements with the specified class
    elements = soup.find_all('h2', {'class':'companyCardWrapper__companyName'})
    
    for h2 in elements:
        # print(h2.getText())
        s = h2.get_text().replace('\t', '').replace('\n', '')
        company__.append(s)

    
```

```python
len(company__)
# company__
```

```python
# File path to save the list items
file_path = 'list_items.txt'

# Open the file in write mode
with open(file_path, 'w') as file:
    # Write each item to a new line in the file
    for item in company__:
        file.write(item + '\n')

print("List items saved to", file_path)
```
