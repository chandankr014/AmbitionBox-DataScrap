```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
# Define headers to mimic a browser request
headers ={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
company_names = []
temp_company = []
```

```python

file_path = 'list_items.txt'
# Open the file in read mode
with open(file_path, 'r') as file:
    # Read a single line from the file
    lines = file.readlines()
    
    for line in lines:
        company_names.append(line.strip())
```

```python
M, N = 1000,1500
for i in range(M,N):
    ns = company_names[i].replace(" ","-")
    temp_company.append(ns)
```

```python
company__ = []
```

```python
iter=0
```

```python
cleaned_strings_r = []
actual_slr = []
salary_min = []
salary_max = []
cleaned_strings = []
designation = []
exp_req = []
#revenue, rating, review, 
reviews = []
ratings = []
```

```python
for company_name in temp_company:
    url = f'https://www.ambitionbox.com/salaries/{company_name}-salaries/'
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    divs_with_td = soup.find_all('tr', {'class': "jobProfiles-table__row"})

    # reinitializing to make it blank
    exp = []
    slr = []
    
    # Iterate through the <div> tags and find paragraphs within <td> tags
    for div in divs_with_td:
        td_tags = div.find_all('td', {'class':'left-content'})
        for td in td_tags:
            experience = td.find_all('div', {'class':'card-content'})
            for i in experience:
                s = i.get_text().replace('\t', '').replace('\n', '')
                exp.append(s)
    
    for i in range(len(exp)):
        cleaned_string = exp[i].replace('\xa0','').replace('Experience', '')
        clear_str = re.sub(r'\([^)]*\)', '', cleaned_string)
        cleaned_strings.append(clear_str)
        sec1, sec2 = cleaned_strings[i].split(':')
        designation.append(sec1)
        exp_req.append(sec2)
        company__.append(company_name)
        
    # Iterate through the <div> tags and find paragraphs within <td> tags
    for div in divs_with_td:
        td_tags = div.find_all('td', {'class':'right-content'})
        for td in td_tags:
            salary = td.find_all('div', {'class':'avg-salary'})
            for j in salary:
                s = j.get_text().replace('\t', '').replace('\n', '')
                slr.append(s)
    
    for i in range(len(slr)):
        try:
            s1,s2,s3 = re.findall(r'[\d,.]+', slr[i])
            actual_slr.append(s1)
            salary_min.append(s2)
            salary_max.append(s3)
        except:
            s1,s2,s3 = None, None, None
            actual_slr.append(s1)
            salary_min.append(s2)
            salary_max.append(s3)
        
    
    # scrap=2
    
    iter += 1
    url2 = f'https://www.ambitionbox.com/list-of-companies?page={iter}'
    response = requests.get(url2, headers=headers)
    soup2 = BeautifulSoup(response.content, 'html.parser')
        
    rt = soup2.find_all('div', class_='companyCardWrapper__companyRatingWrapper')
    for span in rt:
        try:
            r = (span.text).strip()
        except:
            r = None
    div_element = soup2.find_all('div', class_='companyCardWrapper__tertiaryInformation')
    for divs in div_element:
        try:
            span_element = divs.find('span', class_='companyCardWrapper__ActionCount')
        except:
            span_element = None
    # appending rewiew and rating
    for k in range(len(exp)):
        ratings.append(r)
        reviews.append(span_element.text)

```
